```python
class listenanalyse():
  def analyse_liste(self, zahlen):
    summe_gerade = 0
    summe_ungerade = 0
    mittelwert_ungerade = 0
    counter_ungerade = 0
    mittelwert_liste = 0
    for x in zahlen:
      if x % 2 == 0:
        summe_gerade += x
      else:
        summe_ungerade += x
        counter_ungerade += 1
    mittelwert_liste = float((summe_gerade + summe_ungerade) / len(zahlen))
    mittelwert_ungerade = float(summe_ungerade / counter_ungerade)
    abstand = 0
    for x in zahlen:
      if abs(x-mittelwert_liste) > abstand:
        abstand = abs(x-mittelwert_liste)
    print("Summe gerade: %d" % summe_gerade)
    print("Mittelwert ungerade: %.2d" % mittelwert_ungerade)
    print("Mittelwert Liste: %.2d" % mittelwert_liste)
    print("Größter Abstand zum Mittelwert der Liste: %.2d" % abstand)
    #print(f"Summe gerade: {summe_gerade}")
    #print(f"Mittelwert ungerade: {mittelwert_ungerade:.2f}")
    #print(f"Mittelwert Liste: {mittelwert_liste:.2f}")
    #print(f"Größter Abstand: {abstand:.2f}")

analyseobjekt = listenanalyse()
analyseobjekt.analyse_liste([2, 3, 8, 10, 2348, -100, 4827, 327, 193, 0, 1])










```
