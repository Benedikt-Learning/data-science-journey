```python
name = input("Gib deinen Namen ein:")
note_mathe = int(input("Was ist deine Mathenote?"))
note_deutsch = int(input("Was ist deine Deutschnote?"))
note_englisch = int(input("Was ist deine Englischnote?"))
durchschnitt = (note_mathe + note_deutsch + note_englisch) / 3

if durchschnitt >= 90:
    print("Sehr gut, dein Durschnittswert liegt bei %s" %durchschnitt)
elif durchschnitt >= 75 and durchschnitt <= 89:
    print("Gut, dein Durschnittswert liegt bei %s" %durchschnitt)
elif durchschnitt >= 60 and durchschnitt <= 74:
    print("Befriedigend, dein Durschnittswert liegt bei %s" %durchschnitt)
else:
  print("Durchgefallen, dein Durschnittswert liegt bei %s" %durchschnitt)


```
