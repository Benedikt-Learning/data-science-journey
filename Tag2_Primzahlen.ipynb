```python
import math
def ist_primzahl():
  n = int(input("Gib eine Zahl ein:"))
  if n < 2:
    print("False")
    return False
  for x in range(2, int(math.sqrt(n) + 1)):
    if n % x == 0:
      if n != x:
        print("False")
        return False

  print("True")

def alle_primzahlen_bis_n():
  n = int(input("Gib eine Grenze ein"))
  prim = []
  isPrime = True
  if n < 2:
    print(prim)
    return False
  for y in range(2,n):
    isPrime = True
    grenze = int(math.sqrt(y) + 1)
    for x in range(2, grenze):
      if y % x == 0:
        if y != x:
          isPrime = False
    if isPrime == True:
      prim.append(y)

  print(prim)


alle_primzahlen_bis_n()









```
