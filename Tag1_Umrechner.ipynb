```python
print ("Willkommen bei meinem Umrechner!")
umrechnung = input("Was möchtest du umrechnen? Kilometer, Meilen, Celsius oder Fahrenheit")
umrechnung = umrechnung.upper()
print(umrechnung)
temp = float(input("Wie viele %s möchtest du umrechnen?" %umrechnung))
if umrechnung == "KILOMETER" or umrechnung == "MEILEN":
  distanz = temp
  if umrechnung == "KILOMETER":
    miles = distanz * 0.62137119
    print("%.2f %s sind %.2f Meilen" % (temp, umrechnung, miles))
  else:
    kilometer = distanz * 1.609344
    print("%.2f %s sind %.2f Kilometer" % (temp, umrechnung, kilometer))
else:
  temperatur = temp
  if umrechnung == "Celsius":
    fahrenheit = temperatur * 33.8
    print("%.2f Grad %s sind %.2f Fahrenheit" % (temp, umrechnung, fahrenheit))
  else:
    celsius = temperatur * -17.222222
    print("%.2f %s sind %.2f Grad Celsius" % (temp, umrechnung, celsius))



```
